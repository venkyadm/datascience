# Object Oriented plotting

```python
import math
```

```python
x=np.arange(0,math.pi*2,0.05)
x # 1-diamension array start 0 end around 7 and step will be 0.05
```

```python
y=np.sin(x)
y
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
## Creating blank or empt canvas
fig = plt.figure()
##Adding axes
ax = fig.add_axes([0.5,0.5,0.5,0.5])
## labeling the axe
ax.set_title('Sine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.plot(x,y)
plt.show()
```

```python
# Cos

x=np.arange(0,math.pi*2,0.05)
y=np.cos(x)
## Creating blank or empt canvas
fig = plt.figure()
##Adding axes
ax = fig.add_axes([0.5,0.5,0.5,0.5])
## labeling the axe
ax.set_title('Cosine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Cos')
ax.plot(x,y)
plt.show()
```

```python
## Command layouts plots
# Subplots will generate empty canvas as well as figure here 4 plots in empty canvas

fig,a=plt.subplots(2,2)  
x = np.arange(1,6) 
a[0][0].plot(x,x*x)
a[0][0].set_title("First plot")
a[0][1].hist(x)
a[0][1].set_title("Histogram")
a[1][0].plot(x,np.sqrt(x))
a[1][0].set_title('Square Root')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('log')
plt.show()
```
