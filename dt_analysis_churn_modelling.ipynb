# Data Visualization

```python
gender_value_count = df_data['Gender'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
gender_value_count
```

```python
gender_value_count = df_data['Gender'].value_counts()
labels = ["Male","Female"]
colors=['yellowgreen','lightskyblue']
#explode = [0.1,0]
plt.pie(gender_value_count,labels=labels,autopct='%.2f')
plt.title("Customer by Gender")
plt.show()
```

```python
customer_by_location = df_data['Geography'].value_counts()
labels = ['France','Germany','Spain']
plt.pie(customer_by_location,colors=['green','orange','yellow'],labels=labels,explode=[0,0.1,0.1],autopct="%.1f")
plt.title("Customer by Geographical location")
plt.show()
```

```python
customer_status = df_data['Exited'].value_counts()
labels = ['False','True']
plt.pie(customer_status, colors=['green','red'],labels=labels,autopct="%.2f")
plt.title("Customer Exit Status")
plt.show()
```
