```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Getting the builtin datasets of seaborns
sns.get_dataset_names()
```

```python
## Line plot
#loading inbuild dataset 'iris'

data = sns.load_dataset("iris")
```

```python
data.info()
```

```python
### Line plot as the data is continous
## line plot is not that much informative
sns.lineplot(x='sepal_length', y ="sepal_width",data=data)
plt.show()
```

```python
### Line plot as the data is continous
## line plot is not that much informative
#hue parameter will be alwasy categorical
sns.lineplot(x='sepal_length', y ="sepal_width",data=data,hue='species')
plt.show()
```

```python
data.head()
```

```python
## Barplot is basically used to aggreagate the categorical data
## according to some methonds and by default its the mean
##
sns.barplot(x='species',y='sepal_length',data=data)
```

```python
## Count Plot bascially counts the categories and return a 
# count of their occurrences.
sns.countplot(x='species',data=data)
plt.show()

```

```python

sns.catplot(x='species',data=data)
plt.show()
```

```python
## Box plot
##Boxplot is sometime known as the box and whisker plot. It shows the
#distribution of the quantitative data
# That represents the comparisons between variables , boxplot show thes
#quariles of the dataset while ahe whiskers extend to
#show the rest of the distribution ie the data indicating the presence
#of the outliers
## Box plot is used for continous data
## 
sns.boxplot(x='sepal_width',data=data)
plt.show()
```

```python
#checking outliers in each category with respect to sepal width
sns.boxplot(x='species',y='sepal_width',data=data)
plt.show()
```

```python
## Violinplot
## It is similar to the boxplot except thhat it provides a higher
# more advanced visualization and uses the kernel
#density estimation to give a better description about the data distribution.

sns.violinplot(x='species',y='sepal_width',data=data)
plt.show()
```

```python
##Stripplt
## It basically creates a scatter plot based on the category.
# It is created using the stripplot
sns.stripplot(x='species',y='petal_width',data=data)
plt.show()
# setosa can be found with petal_width o to 1
#Versicolor can be found with petal width between 1-2
```

```python
#Histogram
## it valid to only for continous variable
## A histogram is basically used to represent data provided in a form of some groups.
## It is accurate method for the graphical respresentation of numberical data distribution

sns.histplot(x='petal_width',data=data)
plt.show()
```

```python
## Hue parameter uses categorical
sns.histplot(x='petal_width',hue='species',data=data)
plt.show()
```

```python
## kde is approximation of historgram to density curve.
##kde=Faluse it will give histplot
sns.histplot(x='petal_width',hue='species',data=data,kde=True)
```

```python
#Distplot
## Distplot is used basically for univariant set of observations and visualize
## i.e  only one observation and hence we choose one particular column of the dataset
## density means the probability of given point to lie on the curve.
sns.distplot(data['petal_width'],hist=False,color='r')
plt.show()
```

```python
#Pairplot
## Pairplot represents pairwise relation acreoss the entire dataframe and supports an additional arguments 
#called hue for categorical separation.
## what it does basically is create a jointplot between every possible numberical
##column and atakes a while if the dataframe is really huge



sns.pairplot(data=data,hue='species')
plt.show()
```
