## Typecasting
* Hierarchy->bool->int->float->str
* Auto
* Forced

* Typecasting is the method to convert one datatype to another.
* There are 2 types of typecasting,
* 1)Autotype casting
* 2)Forced Typecasting

```python
# Auto Typecasting
exp = 5 + False + 6.78 + True + 7 + 98.34   # 5.0 + 0.0 + 6.78 + 1.0 +7.0 + 98.34
exp
```

```python
type(exp) ## if float is present in any expression,the datatype of entire
## expression will be float only.
```
