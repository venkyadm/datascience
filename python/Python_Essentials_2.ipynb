# Lambda Function

```python
A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.
```

```python
# Syntax
lambda arguments : expression
```

```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
def myfunc(n):
    return lambda a : a * n

mydoubler = myfunc(2)

```

```python
print(mydoubler(11))
```
